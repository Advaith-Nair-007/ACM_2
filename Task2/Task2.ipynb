```python
#importing lib and loading the csv file
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
report = pd.read_csv(r"C:\Users\ASUS\Downloads\StudentsPerformance.csv")
report.head(10)
```

```python
#Trying to understand the data

report.info()
report.describe()
report.isnull().sum()
```

```python
#Ploting histogram to get an idea about the distribution

plt.figure(figsize=(6,4))
plt.hist(report['math score'], bins=20)
plt.title("Distribution of Exam Scores")
plt.xlabel("Math Score")
plt.ylabel("Number of Students")
plt.show()
```

```python
#Plotting a box graph trying to know the outlier
#Visualization 2
sns.boxplot(x='gender', y='math score', data=report)
plt.title("Math Score by Gender")
plt.show()
```

```python
#Visualization 3
sns.boxplot(x='test preparation course', y='math score', data=report)
plt.title("Effect of Test Preparation on Math Score")
plt.show()
```

```python
#Visualization 4
plt.figure(figsize=(10,5))
sns.boxplot(x='parental level of education', y='reading score', data=report)
plt.xticks(rotation=45)
plt.title("Reading Score vs Parental Education")
plt.show()
```

```python
#Visualization 5
sns.heatmap(report[['math score', 'reading score', 'writing score']].corr(), annot=True)
plt.title("Correlation Between Subjects")
plt.show()
```

```python
#Visualization 6
sns.boxplot(x='lunch', y='math score', data=report)
plt.title("Lunch Type vs Math Performance")
plt.show()
```
