```python
x['TotalCharges'] = pd.to_numeric(x['TotalCharges'], errors='coerce')
```

```python
x.isnull().sum()
```

```python
x['TotalCharges'].fillna(x['TotalCharges'].median(), inplace=True)
```

```python
x.isnull().sum()
```

```python
x.drop(columns=['customerID'], inplace=True)
```

```python
q1 = x['MonthlyCharges'].quantile(0.25)
q3 = x['MonthlyCharges'].quantile(0.75)
IQR = q3 - q1

lower = q1 - 1.5 * IQR
upper = q3 + 1.5 * IQR

x = x[(x['MonthlyCharges'] >= lower) & (x['MonthlyCharges'] <= upper)]

```

```python
x['Churn'] = x['Churn'].map({'Yes': 1, 'No': 0})
```

```python
x = pd.get_dummies(x, drop_first=True)
```
